```python
class MyDeque():
    def __init__(self):
        self.deque= []

    def push_front(self, x):
        l = [x]
        self.deque=l+self.deque
    
    def push_back(self, x):
        self.deque.append(x)
    
    def pop_front(self):
        if self.empty():
            return -1
        else:
            pop = self.deque[0]
            self.deque = self.deque[1:]
            return pop
    
    def pop_back(self):
        if self.empty():
            return -1
        else:
            pop = self.deque[-1]
            self.deque = self.deque[0:len(self.deque)-1]
            return pop
    
    def size(self):
        return len(self.deque)

    def empty(self):
        if len(self.deque) == 0:
            return 1
        return 0

    def front(self):
        if self.empty():
            return -1
        else:
            return self.deque[0]
    
    def back(self):
        if self.empty():
            return -1
        else:
            return self.deque[-1]
    
    def __str__(self):
        return "mydeque("+str(self.deque[::1])+")"
    
    def __repr__(self):
        return "mydeque("+str(self.deque[::1])+")"

def main():

    mydeque = MyDeque()

    #push를 이용하여 [3,4,5,6] 만들기
    mydeque.push_back(5)
    mydeque.push_front(4)
    mydeque.push_back(6)
    mydeque.push_front(3)

    print(mydeque)

    print(mydeque.pop_front())
    print(mydeque.pop_back())
    print(mydeque)

    print(mydeque.size())
    print(mydeque.empty())

    print(mydeque.front())
    print(mydeque.back())

    print(mydeque)


if __name__ == "__main__":
    main()
```
