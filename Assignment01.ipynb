```python
#for과 range를 사용해서 1000부터 9999까지 수를 올려가며
#s에 문자열로 변환하여 저장시킨 후,
#if문을 사용하여 만약 문자열 s안에 '10'이 포함되어 있다면 출력하도록 작성했다.

for i in range(1000, 10000):
        s = str(i)
        if ('10' in s):
            print(i)
```

```python
#for과 range를 사용해서 10부터 99999까지 수를 올려가며
#만약 숫자가 20으로 나누어 떨어진다면 s에 문자열로 변환하여 저장시킨 후,
#if문을 사용하여 만약 문자열 s안에 '080'이 포함되어 있다면 출력한다.

for i in range(10, 100000):
    if (i%20==0):
        s = str(i)
        if ('080' in s):
            print(i)
```

```python
#풀이과정
#1. 첫번째 이동 (9:30분 이동)은 무조건 1,3,5,7 중 하나에 도착한다 (Home:4에서 출발하기 때문)
#2. 리스트를 이용하여 각 숫자에 도착했을 때 이동할 수 있는 숫자를 적어놓는다.
#{예} 0번에 도착했다면 리스트[0]에는 [1,3]을 적어서 1번 혹은 3번으로 이동할 수 있도록 한다.
#여기서 이동은 random안에 choice라는 메소드를 이용한다. 리스트 안에 있는 숫자 중 랜덤으로 하나를 뽑을 수 있다.
#3. 두번째 이동부터는, move라는 리스트에서 현재위치(x)에 해당하는 리스트를 뽑아서 그 안에 숫자 중 한 곳으로 이동한다.
#4. 9:00부터 18:00까지는 총 18번의 움직임이 있기 때문에 첫 이동을 뺀 17번의 이동을 move리스트에서 뽑아 이동시킨다.
#5. 이동이 끝나면 현재 위치를 리스트 result에 해당 번호를 1증가시킨다.
#6. 그렇게 10000번을 반복한 후 result리스트를 반복 횟수(repeat)만큼 나눠서 소숫점 이하 4자리수까지 각각의 번호에 있을 확률과
#7. Hospital에 있을 확률, 즉(result[0]/repeat)을 출력한다.



import random

d = {'Hospital':0, 'PostOffice':1, 'Phamacy':2, 'School':3, 'Home':4, 'Convenience':5, 'DepartmentStore':6, 'BeautySalon':7, 'Lotteria':8}
    
first_move = [1,3,5,7]
move = [[1,3], [0,2,4], [1,5], [0,4,6], [1,3,5,7], [2,4,8], [3,7], [4,6,8], [5,7]]
result = [0,0,0,0,0,0,0,0,0]
repeat = 10000
count = 0
print('----------------출발---------------')
    
while(count < repeat):
    x = random.choice(first_move)
    for i in range(17):
        x = random.choice(move[x])
    result[x] += 1
    count += 1
    
for p in range(0, 9):
    print('{:.4f}'.format(result[p]/repeat))
    
print('Hospital에 있을 확률 : {:.4f}'.format(result[0]/repeat))
```
